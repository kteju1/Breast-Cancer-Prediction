```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
Breast_Cancer_Dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(Breast_Cancer_Dataset)
```

```python
df=pd.DataFrame(Breast_Cancer_Dataset.data,columns=Breast_Cancer_Dataset.feature_names)
```

```python
df['label']=Breast_Cancer_Dataset.target
```

```python
X=df.drop(columns=['label'],axis=1)
Y=df['label']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

**Why Standardize**?
Standardization ensures that all features have the same scale (mean = 0, standard deviation = 1), which helps:

Improve model convergence in gradient-based models (e.g., logistic regression, neural networks).
Prevent features with large values from dominating those with smaller values.
Enhance performance in distance-based models (e.g., KNN, SVM).

**What's happening?**
scaler.fit_transform(X_train)

fit() computes the mean and standard deviation from X_train.
transform() applies the standardization:
Xstd=X-mean/std

This ensures that X_train_std has a mean of 0 and standard deviation of 1.
scaler.transform(X_test)

Uses the same mean and standard deviation from X_train (computed earlier).
Ensures X_test is standardized in the same way as X_train

 **Why Not fit_transform(X_test)?**
If we did fit_transform(X_test), it would compute a different mean and std dev for X_test, which can lead to inconsistencies between training and testing data.
Always fit on training data and transform both train & test to ensure proper scaling.

```python
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

Building Neural Network using Tensorflow and Keras

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up layers of NN
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
    ])
```

```python
# compiling the NN

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the NN
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=100)
```

```python
#Visualizing the training history
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['train','val'],loc='upper left')
plt.show()
```

checking accuracy of Test data

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```

```python
Y_pred=model.predict(X_test_std)
```
